#### 2) Encontre a solução geral para y'= -2y+5


$\frac{dy}{dx}= -2y +5$

$\frac{dy}{dx}= -2\cdot(y - 2,5)$ 

$\frac{dy}{y - 2,5}= -2\cdot(dx)$ 

operando a substituição de variável:

$ 
u = y - 2,5
\\
du = dy
$ 

assim: 
$
\int \frac{du}{u}= \int -2 \cdot dx
$

$ ln |u| = - 2x + c_1$

sendo $c_1$ uma constante. 

$ln | 2y - 2,5 | = -2x+ c_1$

$e^{ln|y-2,5|}= e^{-2x + c_1}$

$e^{ln|y-2,5|}= e^{-2x}\cdot e^{ c_1} \Rightarrow  e^{-2x}\cdot C  $ 

Assim, a família de soluções é:



$y = e^{-2x}\cdot C - 2,5$




#### 3) Encontre a solução geral de y'=(sec(y)ˆ2)


$\frac{dy}{dx}=sec^2(y)$

$sec^2(y)= \frac{1}{cos^2(y)}$

$cos^2(y)dy = dx$ $\Rightarrow$ $\int cos^2(y)dy = \int dx$


Tem-se que $cos^2(y)=\frac{1+cos(2y)}{2}$ 


$\int \frac{[1+cos(2y)]dy}{2} = x+C$

$\frac{1}{2}\int [1+cos(2y)]dy = x+C$

$\frac{1}{2}[\int dy +\int cos(2y)] = x+C$

$\frac{y}{2}+\frac{1}{2} \int cos(2y)= x+C$



$u=2y$ 

assim: $dy=du/2$, portanto $\int \frac{cos u\cdot du}{2}=\frac{sen (u)}{2}+c$

ao voltar para a variável y, temos: $\int cos(2y)=\frac{sen (2y)}{2}+C$

Assim, é obtida a solução implícita $\frac{y}{2}+\frac{1}{4} sen(2y)= x+C$
 



#### 4)  Questão ano 2020 (Difícil): Uma população de ratos segue a seguinte $ED p'=0.5*p-450$, onde p é a população de ratos (unidades) em função do tempo (meses). Encontre em quanto tempo a população se torna extinta se a população inicial é de 850 unidades. E para a população inicial de 950, há risco de extinção?

$p'=0,5\cdot t-450$

$\frac{dp}{dt}= 0,5p -450$

$\frac{dp}{0,5p-450}=dt$

Considerando $u=0,5p - 450$: $du= 0,5dt$

$\frac{du}{u}=dt$, assim:

$2\int{\frac{du}{u}}$= $\int{dt}$

$ 2ln|u|= t + c $, sabendo que a $p\geq 0$, considera-se $t+c$ positivo.

$2 ln (0,5p-450)= t + c$


**a)** Considerando $p_0=850$, em $t=0$, temos: 



$2\cdot ln(0,5\cdot 850 -450) = 0+ c$

$c = ln(-25)$, assim: $2\cdot [ln(0,5P-450)- ln(-25)]= 0,5t$

$ln \left( \dfrac{0,5p-450}{-25}\right)= 0,5t$

Aplicando-se a exponencial:

$\left( \dfrac{0,5p-450}{-25}\right)= e^{0,5t} \Rightarrow p= -50e^{0,5t}+900$ 

Calcula-se o limite com o objetivo de observar se há risco de extinção: 

$$\lim_{t\rightarrow \infty} (-50e^{0,5t}+900) = -\infty$$ 

Então, sabe-se que a tendência de função é ir para $-\infty$, ou seja: Ela passa pelo zero. Há risco de extinção.

Calculamos o tempo que leva para a extição acontecer:
$\Rightarrow t= 2\cdot ln (18) \approx 5,78$ meses para a extinção.




**b)** Considerando $p_0=950$, em $t=0$, temos: 


$2\cdot ln(0,5\cdot 950 -450) = 0+ c$

$c = ln(25)$, assim: $2\cdot [ln(0,5P-450)- ln(25)]= 0,5t$

$ln \left( \dfrac{0,5p-450}{25}\right)= 0,5t$

Aplicando-se a exponencial:

$\left( \dfrac{0,5p-450}{25}\right)= e^{0,5t} \Rightarrow p= 50e^{0,5t}+900$ 

Calcula-se o limite com o objetivo de observar se há risco de extinção: 

$$\lim_{t\rightarrow \infty} (-50e^{0,5t}+900) = +\infty$$ 

Então, sabe-se que a tendência de função é crescer indefinidamente, não há risco de extinção.
